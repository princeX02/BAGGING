```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=1000,noise=0.2)
```

```python
df={"x1":x[:,0],"x2":x[:,1],"y":y}
```

```python
data=pd.DataFrame(df)
```

```python
data.head(2)
```

```python
sns.scatterplot(x="x1",y='x2',data=data,hue='y')
plt.show()
```

```python
x_a=data.iloc[:,:-1]
y_a=data["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_a,y_a,random_state=42,test_size=0.2)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg=BaggingClassifier(estimator=SVC() ,n_estimators=30)
bg.fit(x_train,y_train)
bg.score(x_train,y_train)*100,bg.score(x_test,y_test)*100
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf.score(x_train,y_train)*100,rf.score(x_train,y_train)*100
```
